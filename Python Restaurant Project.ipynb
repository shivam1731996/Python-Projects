```python
menu = {
    'pizza': 40,
    'pasta': 50,
    'burger': 60,
    'salad': 70,
    'coffee': 80,
    'sandwich': 45,
    'fries': 35,
    'noodles': 55,
    'wrap': 50,
    'ice cream': 60,
    'tea': 20,
    'cold drink': 30,
    'juice': 40,
    'milkshake': 70
}

print(" Welcome to Python Restaurant!")
print("Here's our menu:\n")
for item, price in menu.items():
    print(f"{item.capitalize():<15} Rs{price}")

order = {}  # To store item: quantity
order_total = 0  # Base amount before tax/discount

# Ordering loop
while True:
    item = input("\nEnter the name of the item you want to order: ").lower()
    if item in menu:
        try:
            quantity = int(input(f"How many {item}s would you like? "))
            if quantity <= 0:
                print("❗ Quantity must be at least 1.")
                continue
        except ValueError:
            print("❗ Please enter a valid number.")
            continue

        # Add to order
        if item in order:
            order[item] += quantity
        else:
            order[item] = quantity

        order_total += menu[item] * quantity
        print(f" {quantity} {item}(s) added to your order.")
    else:
        print(" Sorry, that item is not on the menu.")

    another = input("Do you want to add another item? (yes/no): ").lower()
    if another != 'yes':
        break

# Tax, discount, delivery charge
tax = 0.05 * order_total  # 5% tax
discount = 0.1 * order_total if order_total > 200 else 0
delivery_charge = 20 if order_total < 150 else 0
final_total = order_total + tax + delivery_charge - discount

# Order summary
print("\n ORDER SUMMARY:")
print(f"{'Item':<15}{'Qty':<5}{'Unit Price':<12}{'Total':<10}")
print("-" * 45)
for item, qty in order.items():
    unit_price = menu[item]
    total_price = unit_price * qty
    print(f"{item.capitalize():<15}{qty:<5}{unit_price:<12}{total_price:<10}")
print("-" * 45)
print(f"{'Subtotal':<32} Rs{order_total:.2f}")
print(f"{'Tax (5%)':<32} Rs{tax:.2f}")
print(f"{'Discount (10%)':<32} -Rs{discount:.2f}")
print(f"{'Delivery Charge':<32} Rs{delivery_charge:.2f}")
print(f"{'Total Amount':<32} Rs{final_total:.2f}")

print("\n Thank you for ordering from Python Restaurant!")
```
